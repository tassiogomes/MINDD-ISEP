```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import folium

from google.colab import drive
drive.mount('/content/drive')

df_clientes = pd.read_csv('drive/My Drive/MINDD/clientes.txt', sep='\t')
df_compras = pd.read_csv('drive/My Drive/MINDD/compras.txt', sep='\t')
df_clientes.head()
```

```python
df_compras.head()
```

```python


# Número de cestas de compras (shopping baskets)


# Número de compras de cada produto numericamente e graficamente
product_purchases = df_compras['ProductID'].value_counts()
product_purchases.plot(kind='bar', figsize=(12, 6))
plt.title('Number of Purchases of Each Product')
plt.xlabel('ProductID')
plt.ylabel('Number of Purchases')
plt.show()

# Número de produtos por cesta de compras
products_per_basket = df_compras.groupby('BasketID')['ProductID'].count()
products_per_basket = products_per_basket.value_counts()

# Contagem de cestas por número de produtos vendidos
baskets_by_product_count = products_per_basket.reset_index()
baskets_by_product_count.columns = ['Products Sold', 'Number of Baskets']

# Exibir informações
print('Number of purchases of each product:')
print(product_purchases)
print('Number of products per shopping basket:')
print(products_per_basket)
print('Count of baskets by number of products sold:')
print(baskets_by_product_count)
```

```python
num_customers = df_clientes['client'].nunique()
num_unique_products = df_compras['product'].nunique()
num_baskets = df_compras['client'].nunique()
product_purchases = df_compras['product'].value_counts()
# histogram
product_purchases.plot(kind='bar', figsize=(6, 2))
plt.title('Number of Purchases of Each Product')
plt.xlabel('ProductID')
plt.ylabel('Number of Purchases')
plt.show()


print(f'Number of customers: {num_customers}')
print(f'Number of different products sold: {num_unique_products}')
print(f'Number of shopping baskets: {num_baskets}')
print('Number of purchases of each product:')
print("----------------------------------------")
print(product_purchases)

```
